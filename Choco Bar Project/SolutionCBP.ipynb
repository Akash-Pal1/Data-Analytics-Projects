```python
list1 = cbdf1['ingredients'].unique()
L = []
for i in list1:
    L.append(i.split(","))
L

flat_list = [item for sublist in L for item in sublist]

list(set(flat_list))
```

```python
q1 = cbdf.drop(['num_ingredients','ingredients','review','bar_name','company_location','manufacturer'],axis=1).groupby("bean_origin")
q1 = q1.agg(np.mean)
q1 = q1['rating'].reset_index(drop=False)
q1
```

```python
q1r = list(q1['rating'])
q1bo = list(q1['bean_origin'])
f = plt.figure()
f.set_figwidth(18)
f.set_figheight(11)
plt.plot(q1bo,q1r,'ko-')
plt.xticks(rotation=90)
plt.figure(figsize=(18,18))
plt.show()
```

```python
bar_list = cbdf1['bar_name'].unique()
L = []
for i in bar_list:
    L.append(i.split(","))
L

bar_name_list = [item for sublist in L for item in sublist]

bar_name_list = list(set(bar_name_list))
len(bar_name_list)
```

```python
q2 = cbdf.groupby("bean_origin")
q2 = q2['bar_name'].agg('count').reset_index(drop=False)
q2 = q2.sort_values(ascending=False,by='bar_name').reset_index(drop=True)
q2 = q2[:10]
q2
```

```python
sns.barplot(data=q2,x='bean_origin',y='bar_name').set(xlabel="Origin of country",ylabel="The number of chocolate bars",title = "Barplot ")
plt.xticks(rotation=270)
plt.show()
```
