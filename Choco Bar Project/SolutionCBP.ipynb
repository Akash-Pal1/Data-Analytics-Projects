### Q2. How many bars were reviewed for each of those countries?

```python
q2 = cbdf.groupby("bean_origin")
q2 = q2['bar_name'].agg('count').reset_index(drop=False)
q2 = q2.sort_values(ascending=False,by='bar_name').reset_index(drop=True)
q2 = q2[:10]
q2
```

```python
sns.barplot(data=q2,x='bean_origin',y='bar_name',palette='Spectral').set(xlabel="Origin of country",ylabel="The number of chocolate bars",title = "Barplot ")
plt.xticks(rotation=270)
plt.show()
```

The above visualization shows the top 10 countries each with the total number of chocolate bars present in thw country which can be viewed in the above dataframe.

### Q3.A. How does cocoa content relate to rating?

```python
cocoa_60_70 = (cbdf[cbdf['cocoa_percent'] > 60 ])
cocoa_60_70 = ((cocoa_60_70[cocoa_60_70['cocoa_percent'] <= 70])[['rating']])['rating']
cocoa_60_70_l = len(cocoa_60_70)
cocoa_60_70 = cocoa_60_70.mean()

cocoa_70_80 = (cbdf[cbdf['cocoa_percent'] > 70 ])
cocoa_70_80 = ((cocoa_70_80[cocoa_70_80['cocoa_percent'] <= 80])[['rating']])['rating']
cocoa_70_80_l = len(cocoa_70_80)
cocoa_70_80 = cocoa_70_80.mean()

cocoa_80_90 = (cbdf[cbdf['cocoa_percent'] > 80 ])
cocoa_80_90 = ((cocoa_80_90[cocoa_80_90['cocoa_percent'] <= 90])[['rating']])['rating']
cocoa_80_90_l = len(cocoa_80_90)
cocoa_80_90 = cocoa_80_90.mean()

cocoa_90 = ((cbdf[cbdf['cocoa_percent'] >  90])[['rating']])['rating']
cocoa_90_l = len(cocoa_90)
cocoa_90 = cocoa_90.mean()


cocoa_0_60 = ((cbdf[cbdf['cocoa_percent'] <=  60])[['rating']])['rating']
cocoa_0_60_l = len(cocoa_0_60)
cocoa_0_60 = cocoa_0_60.mean()

cocoa_0_60
```

```python
l = list([cocoa_0_60,cocoa_60_70,cocoa_70_80,cocoa_80_90,cocoa_90])
ln = list([cocoa_0_60_l,cocoa_60_70_l,cocoa_70_80_l,cocoa_80_90_l,cocoa_90_l])
names = ['< 60',' 60 - 70','70 - 80','80 - 90',' > 90']
sns.barplot(x=names,y=l).set(xlabel="Cocoa percentage interval",ylabel="Average Rating inside interval",title = "Barplot")
plt.show()
print(ln)
```

The various classes have been created by us to show the average rating of the bars based on the cocoa content present in the bars. The various labels are **less than 60%, between 60 and 70, between 70 and 80, between 80 and 90, and greater than 90** as the intervals for classification. 

It is observed that the class **60-70 percent** cocoa content percentage has both the highest rating and highest number of users.

While the bars having more than **90 percent** of cocoa content present have least rating and number of users.
